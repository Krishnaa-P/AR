```python
# Install required libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Age': [22, 24, 26, 28, 30, 32, 34, 36, 38, 40],
    'Gender': ['Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Male', 'Female', 'Female'],
    'Income': [10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000],
    'Education': ['PhD', 'PhD', 'PhD', 'PhD', 'Master', 'PhD', 'Bachelor', ' PhD', 'PhD', 'Bachelor'],
    'Viewership': [2500000, 3000000, 3500000, 4000000, 4500000, 5000000, 5500000, 6000000, 6500000, 7000000]
}
df = pd.DataFrame(data)
print(df.head())
print(df.describe())
sns.histplot(df['Age'], bins=2)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
sns.boxplot(x=df['Income'])
plt.title('Income Distribution')
plt.xlabel('Income')
plt.show()
education_counts = df['Education'].value_counts()
education_counts.plot(kind='barh')
plt.title('Education Level Distribution')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.show()
sns.scatterplot(data=df, x='Age', y='Viewership', hue='Gender')
plt.title('Viewership vs. Age')
plt.xlabel('Age')
plt.ylabel('Viewership')
plt.legend(title='Gender')
plt.show()
```
